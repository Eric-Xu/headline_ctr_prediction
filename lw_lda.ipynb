```python
import pandas as pd
from gensim import corpora, models, similarities

# source: 15_NLP/gensim_demo.ipynb
# source: https://github.com/ga-students/DS-LA-03/blob/master/src/lesson15/gensim_demo.py
```

```python
df = pd.read_csv("datasets/livewire_hits2.csv")
```

```python
df.head()
```

```python
x_train = df.title[0:100]
y_train = df.day_0[0:100]

x_test = df.title[100:200]
y_test = df.day_0[100:200]
```

```python
lda_train = pd.DataFrame(columns = ["title", "day_0_count"])
lda_train.title = x_train
lda_train.day_0_count = y_train
lda_train.head()
```

### Remove common words and tokenize

```python
stoplist = set('for a of the and to in'.split())
```

```python
texts = [[word for word in title.lower().split() if word not in stoplist] for title in lda_train.title]

for t in texts[0:5]:
    print t
```

#### Remove words that appear only once

```python
all_tokens = []
for list in texts:
    all_tokens = all_tokens + list

print len(all_tokens)
all_tokens[0:20]
```

```python
tokens_once = set(word for word in set(all_tokens) if all_tokens.count(word) == 1)

print len(tokens_once)
```

```python
freq_texts = [[word for word in text if word not in tokens_once]
          for text in texts]

print len(freq_texts)
freq_texts[0:20]
```

### Turn each document (headline) into a vectorized bag-of-words

```python
# Creating a dictionary to map words to integers.
dictionary = corpora.Dictionary(freq_texts)
print dictionary
```

```python
print dictionary.token2id.items()[:10]
```

```python
new_doc = "Pope visits NJ mayor"
new_vec = dictionary.doc2bow(new_doc.lower().split())
print new_vec
```

```python
# Vectorize: turn each document (list-of-words) into a vectorized bag-of-words.
corpus = [dictionary.doc2bow(text) for text in texts]

print len(corpus)
corpus[0:20]
```

```python
print "freq_text length: ", len(freq_texts)
print "dictionary length: ", len(dictionary)
print "corpus length: ", len(corpus)
```

### Bag-of-words to Tf-IDF

```python
# Transformations, from one vector representation to another.
# E.g. bag-of-words to Tf-IDF
# step 1 -- initialize a model. This learns document frequencies.
tfidf = models.TfidfModel(corpus)
```

```python
# You can transform a vectorized bag-of-words:
doc_bow = [(0, 1), (1, 1)]
print(tfidf[doc_bow])
```

```python
# You can also transform a whole corpus.
# Note that corpus_tfidf is actually an iterator over corpus
# that performs the transformation at the last-minute.
# step 2 -- use the model to transform bunch of vectors.
corpus_tfidf = tfidf[corpus]
```

```python
for c in corpus_tfidf:
    print c
```

### Latent Sementic Analysis (LSA)

```python
# Now let's try Latent Sementic Analysis (LSA) to transform documents into 2D space.
# initialize an LSI transformation
lsi = models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=10)

# create a double wrapper over the original corpus: bow->tfidf->fold-in-lsi
corpus_lsi = lsi[corpus_tfidf]
```

```python
for c in corpus_lsi:
    print c
```

### Topic modeling with LDA

```python
# extract 10 LDA topics; use the default one-pass algorithm
lda = models.ldamodel.LdaModel(corpus, id2word=dictionary, num_topics=10)
```

```python
# print the most contributing words for 5 randomly selected topics
for topic in lda.print_topics(5):
    print topic
```

### Similarity Queries

```python
query = "jesus mayor pope"

# convert query into bag-of-words
query_bow = dictionary.doc2bow(query.lower().split())
print "query bow ", query_bow

# convert bow into LSI space
query_lsi = lsi[query_bow]
for qlsi in query_lsi:
    print "query lsi ", qlsi

# convert bow into LDA space
query_lda = lda[query_bow]
for qlda in query_lda:
    print "query lda ", qlda
```

### Initializing query structures

```python
# Uses cosine measure
# Transform corpus to LDA space
lda_sim_matrix = similarities.MatrixSimilarity(lda[corpus])
for m in lda_sim_matrix:
    print m
```

```python
# perform a similarity query against the corpus
sims = lda_sim_matrix[query_lda]
sims_sorted = sorted(enumerate(sims), key=lambda item: -item[1])

print "sims length ", len(sims_sorted)
for sim in sims_sorted[0:20]:
    print sim
```
