### Data Munging

```python
def regex_match():
    brackets_match = "\[u|\[|\]"
    dashes_match = "\s\-"
    backslash_match = "\\\\"
    quotes_match = "\'|\""
    em_match = ",\s<em>|<em>|</em>,\su|</em>"
    strong_match = ",\s<strong>|<strong>|</strong>,\su|</strong>"
    i_match = ",\s<i>|<i>|</i>,\su|</i>"
    char_codes = "\&amp\;"
    full_match = brackets_match + "|" + dashes_match + "|" + backslash_match + "|" + quotes_match + "|" + em_match  + "|" + strong_match + "|" + i_match + "|" + char_codes
    # additional chars - . , ; : ! ? 
    return full_match
```

```python
for h in df_raw.Headlines:
    cleaned = re.sub(regex_match(), '', str(h))
    print cleaned
```
